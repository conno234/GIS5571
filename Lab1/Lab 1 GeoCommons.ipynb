```python
#First we will import the needed packages for this notebook
import requests
import urllib3

#Since some of the APIs may require certificate authorization, we'll use the below code to get around that.
urllib3.disable_warnings()
```

```python
#First we will do an API call from the Minnesota Geospatial Commons to get the Minnesota's Counties Boundaries from the 2020 Census.
#Before we make the actual call, we can assign a variable to the dataset's URL, formatted according to CKAN's API commands. 
api_url = r"https://gisdata.mn.gov/api/3/action/package_show?id=us-mn-state-metc-bdry-census2020counties-ctus"
```

```python
#We will use the requests package to make the actual API call and save it was "response".
#"verify=False" disables verifying the URL, meaning a certificate does not have to be produced to call the data.
response = requests.get(api_url, verify=False)
```

```python
response.text()
```

```python
#Next, we will convert the response to a json
json=response.json()
```
